## clean data

* > 1- drop duplicated raw


```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.duplicated().sum()
```

* > 3- remove row from UnitPrice columns < 0


```python
dataset = dataset[dataset['UnitPrice'] > 0]
```

```python
dataset.describe()
```

* > 4- Separate sales data from return data

```python
sales_data = dataset[
    (dataset['Quantity'] > 0) &
    (~dataset['InvoiceNo'].astype(str).str.startswith('C'))
]
sales_data.head()
```

```python
sales_data.shape
```

```python
returns_data = dataset[
    (dataset['Quantity'] < 0) |
    (dataset['InvoiceNo'].astype(str).str.startswith('C'))
]
returns_data.head()
```

```python
sales_data.describe()
```

```python
sales_data[sales_data['UnitPrice'] > sales_data['UnitPrice'].quantile(0.999)]
```

```python
sales_data[sales_data['Description'] == 'DOTCOM POSTAGE']
# Prices vary because within "DOTCOM POSTAGE" POSTAGE there are multiple products, each with a different stock code.
```

```python
sales_data[sales_data['UnitPrice'] == sales_data['UnitPrice'].max()]
# fees was paid to Amazon; this is not the price of the product.
```

```python
sales_data[sales_data['Quantity'] == sales_data['Quantity'].max()]
```

* > 6- remove max value in UnitPrice columns

```python
####
sales_data.drop(sales_data[sales_data['UnitPrice'] == sales_data['UnitPrice'].max()].index, inplace=True)
```

```python
sales_data[sales_data['UnitPrice'] == sales_data['UnitPrice'].max()]
```
