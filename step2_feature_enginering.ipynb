## feature Enginering
* > 1- create total price col.
* > 2- create weekend col.
* > 3- create hour col.
* > 4- create DayOfWeek col.
* > 5- create InvoiceMonth col.
* > 6- create InvoiceYear col.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rfm_data = pd.read_csv('rfm_data.csv')
rfm_data.head()
```

```python
rfm_data.dtypes
```

```python
rfm_data['Total_price'] = rfm_data['Quantity'] * rfm_data['UnitPrice']
rfm_data.head()
```

```python
"""
total_price :
It was created because it is important for the following:

The “Monetary” foundation in RFM
Identifying VIP clients
Profitability analysis

-------------------------------------------------------

InvoiceYear : 
Comparing performance over the years /  analyzing growth or decline.

---------------------------------------------------------

InvoiceMonth :
Discovering Seasonality / Peak Sales Months

---------------------------------------------------------

DayOfWeek:
Understanding weekly buying behavior
Do people buy more on certain days?


-------------------------------------

Hour :
Understanding peak times
Improving marketing campaigns


-----------------------------

Weekend:
Do people shop more on weekends?


"""
```

```python
rfm_data['InvoiceYear'] = rfm_data['InvoiceDate'].dt.year
rfm_data['InvoiceMonth'] = rfm_data['InvoiceDate'].dt.month
rfm_data['DayOfWeek'] = rfm_data['InvoiceDate'].dt.day_name()
rfm_data['hour'] = rfm_data['InvoiceDate'].dt.hour
rfm_data['weekend'] = (rfm_data['InvoiceDate'].dt.weekday >= 5).astype(int)
rfm_data.head()
```

```python
rfm_data.to_csv('rfm_final_data.csv', index=False)
```
